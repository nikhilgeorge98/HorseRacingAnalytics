```python
def point5round(a):
    return math.ceil(a * 2) / 2;
```

```python
def conjunction(type, *conditions):
    if type:
        return functools.reduce(np.logical_or, conditions)
    else:
        return functools.reduce(np.logical_and, conditions)
```
