```python
import numpy as np
import pandas as pd
import datetime
import re
```

```python
# data = pd.concat(pd.read_excel(r'C:\Users\Testing\NEW METHOD FILE.xlsx', sheet_name=None), ignore_index=True)

data = pd.read_excel(r'C:\Users\Testing\Downloads\ALL METHODS - MASTER FILE (1).xlsb', engine='pyxlsb')
data['Date']= pd.to_datetime(data['Date'], unit='D', origin='1899-12-30T00:00:00.000000000')
data.rename(columns = {'CENTER':'Centre', 'VENUE ID': 'Venue ID', 'R No #': 'Race No', 'RES': 'Result', 'CLASS': 'Class', 'DIST': 'Dist'}, inplace = True)
data.Dist.replace([np.nan],[0],inplace=True)
data.Class.replace([np.nan],['-'],inplace=True)
data.Jockey.replace([np.nan],['-'],inplace=True)
data.Trainer.replace([np.nan],['-'],inplace=True)
data.LBW.replace(['W'],[100],inplace=True)
data['Horse Name'].replace([np.nan],['-'],inplace=True)
data['Sr #'] = [x+1 for x in data.index]
data['Rtg'] = [0]*len(data)
data['Penalty'] = [0]*len(data)
data['Time in S'] = [0]*len(data)
data['Time'] = [0]*len(data)
data['Odds'] = [0]*len(data)

data.rename(columns = {'Center':'Season', 'Horse Name': 'Horse', 'Dist': 'Distance', }, inplace = True)

data.Age.replace([np.nan, '3y', '5y', '4y', '6y', '7y', '8y', '9y', '2y', '0y', '14y', '10y', '11y', '12y', '1y'],[0, 3, 5, 4, 6, 7, 8, 9, 2, 0, 14, 10, 11, 12, 1],inplace=True)
data.Result.replace(['DNC', 'NC', 'DNF', 'DR', 'W'],[0,0,0,0,0], inplace=True)
data.Sh.replace(['s', 'a', np.nan], ['S', 'A', 'A'], inplace=True)

data['RunNo.']=[-1]*len(data)

data['Dr'].replace([np.nan], [0], inplace=True)
data['Dr'] = data.Dr.astype(int)

data.drop(columns=['Time'], inplace=True)
data['Month'] = data['Date'].dt.month
mylist = data.Class.unique()
r = re.compile("^IV|^V")
allowed_classes = list(filter(r.match, mylist))
allowed_months = [5,6,7,8,9]
data = data.loc[data['Month'].isin(allowed_months) & data['Class'].isin(allowed_classes)]
```

```python
df=data
df
```

```python
race_stats = df.groupby(["Date", "Season", "Race No", "Result"], as_index=True).size().unstack(
    fill_value=0)
race_stats.reset_index(inplace=True)
race_stats.index = np.arange(1, len(race_stats) + 1)
race_stats['min_horses']= race_stats[9]==1
race_stats = race_stats[['Date', 'Season', 'Race No', 'min_horses']]
race_stats
```

```python
mylist = data.Class.unique()
r = re.compile("(.+?)4y&o$")
filtered_allowed_classes = list(filter(r.match, mylist))
qualified_horses = df.loc[(df['Class'].isin(filtered_allowed_classes)) & (df['LBW']>=4) & (df['Result'].isin([2,3,4])) & (df['Age']>=5)][['Horse', 'Date', 'Season', 'Race No', 'Sr #', 'Trainer', 'Rtg', 'Class', 'Distance','Age',
 'Wt', 'Jockey', 'Dr', 'Sh', 'Result', 'LBW', 'Penalty', 'Time in S', 'RunNo.']]
qualified_horses.index = np.arange(1, len(qualified_horses) + 1)
qualified_horses = qualified_horses.merge(race_stats,indicator = True, how='left')
qualified_horses = qualified_horses.loc[qualified_horses['min_horses']==True]
qualified_horses.index = np.arange(1, len(qualified_horses) + 1)
qualified_horses
```
