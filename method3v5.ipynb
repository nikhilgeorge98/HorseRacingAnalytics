```python
# data = pd.concat(pd.read_excel(r'C:\Users\Testing\NEW METHOD FILE.xlsx', sheet_name=None), ignore_index=True)
# data = pd.read_excel(r'C:\Users\Testing\Downloads\NEW METHOD 210822 (1).xlsx')
# data.dropna(subset=['Sr #'], inplace=True)###########

data = pd.read_excel(r'C:\Users\Testing\Downloads\ALL METHODS - MASTER FILE (1).xlsb', engine='pyxlsb')
data['Date']= pd.to_datetime(data['Date'], unit='D', origin='1899-12-30T00:00:00.000000000')
data.rename(columns = {'CENTER':'Centre', 'VENUE ID': 'Venue ID', 'R No #': 'Race No', 'RES': 'Result', 'CLASS': 'Class', 'DIST': 'Dist'}, inplace = True)
data.Dist.replace([np.nan],[0],inplace=True)
data.Class.replace([np.nan],['-'],inplace=True)
data.Jockey.replace([np.nan],['-'],inplace=True)
data.Trainer.replace([np.nan],['-'],inplace=True)
data['Horse Name'].replace([np.nan],['-'],inplace=True)
data['Sr #'] = [x+1 for x in data.index]
data['Rtg'] = [0]*len(data)
data['Penalty'] = [0]*len(data)
data['Time in S'] = [0]*len(data)
data['Time'] = [0]*len(data)
data['Odds'] = [0]*len(data)

data.rename(columns = {'Center':'Season', 'Horse Name': 'Horse', 'Dist': 'Distance', }, inplace = True)

data.Age.replace([np.nan, '3y', '5y', '4y', '6y', '7y', '8y', '9y', '2y', '0y', '14y', '10y', '11y', '12y', '1y'],[0, 3, 5, 4, 6, 7, 8, 9, 2, 0, 14, 10, 11, 12, 1],inplace=True)
data.Result.replace(['DNC', 'NC', 'DNF', 'DR', 'W'],[0,0,0,0,0], inplace=True)
data.Jockey.replace([np.nan], [""],inplace=True)
data.Trainer.replace([np.nan], [""],inplace=True)
data['LBW'].replace([np.nan, 'W'], [0,0], inplace=True)
data['Sh'].replace(['s', 'a', np.nan], ['S','A','-'], inplace=True)

data['combined_data'] = list(zip(data.Age, data.Result))
data['LBW2'] = ((data['Result']==2).replace([False, True], [0, 1]))*data['LBW']
# data['LBW2'].replace([''], [], inplace=True)

data['RunNo.']=[-1]*len(data)

data['Dr'].replace([np.nan], [0], inplace=True)
data['Dr'] = data.Dr.astype(int)
data['Date'] = data['Date'].dt.date

data.drop(columns=['Time'], inplace=True)
###########################
new = data['Venue ID'].str.split("=", n = 1, expand = True)
data['Venue ID'] = new[1]
data['Venue ID'] = data['Venue ID'].astype(int)
data['Venue Name'] = data['Venue ID'].replace([1,2,3,4,7,8,9,10,11], ['KOL', 'MUM', 'BLR', 'CHE', 'DEL', 'MYS', 'OOT', 'PUN', 'HYD'])
# data['Sr #'] = data['Sr #'].astype(int)
# data['Distance'] = data['Distance'].astype(int)
# data['Race No'] = data['Race No'].astype(int)
```

```python
combo=[(3, 1), (5,2)]
m = df.groupby(["Date", "Venue ID","Season", "Race No", "combined_data"], as_index=True).size().unstack(
    fill_value=0)
mis_col = [i for i in combo if i not in m.columns]
for col in mis_col:
    m[col] = [0] * len(m)
m = m[combo]
n = df.groupby(["Date", "Venue ID","Season", "Race No"], as_index=True)['LBW2'].sum()
l=df.groupby(["Date", "Venue ID","Season", "Race No","Result"], as_index=True).size().unstack(fill_value=0)
race_stats = pd.concat([m, n, l], axis=1)
race_stats['LBW2'] = race_stats['LBW2']/race_stats[2]
race_stats.reset_index(inplace=True)
race_stats = race_stats[["Date", "Venue ID","Season", "Race No"] + combo + ["LBW2"]]
race_stats.index = np.arange(1, len(race_stats) + 1)
race_stats['initial_conditions'] = race_stats[(3,1)]*race_stats[(5,2)] & (race_stats['LBW2'] <= 0.75)
race_stats
```

```python
qualifying_races = race_stats.loc[race_stats['initial_conditions']==1]
qualifying_races = qualifying_races[["Date", "Season", "Race No","Venue ID"]]
qualifying_races
```
